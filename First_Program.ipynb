```python
def median_imputation(input_df):
  numeric_cols = input_df.select_dtypes(include=[np.number]).columns
  for col in numeric_cols:
    median_value = input_df[col].median()
    input_df[col].fillna(median_value, inplace=True)
  return input_df

input_df = median_imputation(input_df)
print(input_df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for column in input_df.columns:
  if input_df[column].dtype == 'object':
    input_df[column] = label_encoder.fit_transform(input_df[column])

print(input_df.head())
```

```python
input_df.isnull().sum()
```

```python
X = input_df.iloc[:, :-1]
y = input_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X_train,y_train)
```

```python
LinReg.score(X_test,y_test)
```
